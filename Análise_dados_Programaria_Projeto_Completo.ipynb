```python
# Ela está calculando o 1º quartil (Q1) da coluna 'SALARIO' do DataFrame dados.
Q1 = dados['SALARIO'].quantile(0.25)
Q1
```

Os quartis dividem os seus dados em quatro partes iguais:

Nome Percentual Significado

Q1 (1º quartil) 25% 25% dos valores estão abaixo dele

Q2 (mediana) 50% 50% dos valores estão abaixo dele

Q3 (3º quartil) 75% 75% dos valores estão abaixo dele

O .25 representa 25% (ou seja, 1/4) da distribuição dos dados.

```python
# Calcular o 3º quartil
# O Q3 separa 3/4 dos dados 0.75
Q3 = dados['SALARIO'].quantile(0.75)
Q3
```

```python
#Interquartil =  efine o valor que entre o 1º e o 3º quartil
IQR = Q3 - Q1
IQR
```

```python
# Buscar os limites. O limite superior é 3 quartis , mais 1.5 vezes o interquartil

limite_superior = Q3 + (1.5*IQR)
limite_superior
```

```python
# O limite inferior é 1 Q 3º quartil memos 1.5 vezes o interquartil
limite_inferior = Q1 - (1.5*IQR)
limite_inferior
```

```python
dados['FAIXA SALARIAL'].value_counts()
```

```python
#Média do salário
media_salario = dados['SALARIO'].mean()
media_salario
```

```python
desvio_salario = dados['SALARIO'].std()
desvio_salario
```

```python
#calculando um limite superior com base na média e no desvio padrão dos salários
limite_superior = media_salario + (3*desvio_salario)
limite_superior
```

media_salario: é a média dos salários (ou seja, o valor médio da coluna SALARIO)

desvio: é o desvio padrão dessa mesma coluna (indica o quanto os salários variam da média)

3 * desvio: está aplicando a regra dos 3 sigmas

🧠 O que significa esse cálculo?

Ele está criando um limite acima da média, usado normalmente para identificar valores muito altos, ou seja, possíveis outliers (valores fora do padrão).

📏 Regra dos 3 Sigmas (ou 3 desvios) É uma regra da estatística que diz:

Em uma distribuição normal, cerca de 99,7% dos dados ficam entre a média - 3 desvios e a média + 3 desvios.

📊 Então, o que esse código faz?

💡 Ele define o ponto máximo que consideramos "normal" para o salário. Se algum salário estiver acima desse limite, pode ser considerado um outlier (valor fora do comum).

```python
dados[dados['SALARIO'] > limite_superior]['FAIXA SALARIAL'].value_counts()
```

🧠 O que está acontecendo? Esse código está buscando quantas pessoas estão com salário acima de um determinado valor (limite_superior) e quantas vezes cada faixa salarial aparece entre essas pessoas.

Vamos por partes:

🪜 Passo a passo:

dados['SALARIO'] > limite_superior 👉 Isso cria uma condição booleana, que verifica quais registros (linhas) têm salário maior que o valor armazenado em limite_superior.

dados[dados['SALARIO'] > limite_superior] 👉 Isso filtra o DataFrame, mostrando apenas as linhas em que o salário é maior que o limite.

['FAIXA SALARIAL'] 👉 Depois de filtrar, ele pega apenas a coluna 'FAIXA SALARIAL' dessas linhas.

.value_counts() 👉 Por fim, ele conta quantas vezes cada faixa salarial aparece nesse grupo filtrado.

```python
media_30_40 = dados[(dados['FAIXA SALARIAL']== 'de R$ 30.001/mês a R$ 40.000/mês') & (dados['SALARIO'] < limite_superior)]['SALARIO'].mean()
```

🧠 O que ele faz? Esse código está calculando a média dos salários reais das pessoas que:

Estão na faixa salarial de R 30.001aR  40.000;

Mas ganham de verdade um valor abaixo do limite superior (ex: fora do grupo de salários muito altos/outliers).

Ou seja, está analisando qual é o valor médio de salário real das pessoas que deveriam estar nessa faixa salarial, mas sem contar os casos extremos.

🪜 Passo a passo: dados['FAIXA SALARIAL'] == 'de R 30.001/mêsaR  40.000/mês' 👉 Filtra só as pessoas que estão nessa faixa salarial declarada.

dados['SALARIO'] < limite_superior 👉 Garante que o salário real não é um outlier (não está acima do limite considerado aceitável para a análise).

& 👉 Une as duas condições (ou seja, ambas precisam ser verdadeiras).

dados[...]['SALARIO'] 👉 Depois de aplicar os dois filtros acima, ele pega a coluna SALARIO dessas pessoas.

.mean() 👉 Calcula a média dos salários reais desse grupo filtrado.

🧩 Qual o objetivo? Você está respondendo a uma pergunta tipo:

"Entre as pessoas que estão na faixa salarial de R 30milaR  40 mil, qual é a média real do salário delas, sem considerar os que têm salários fora do comum (outliers)?"

Esse tipo de análise ajuda a entender se as pessoas estão realmente ganhando dentro da faixa que declaram ou se há uma diferença importante ali — às vezes, pode até indicar inconsistência nos dados!

```python
dados.loc[
    (dados['FAIXA SALARIAL'] == 'de R$ 30.001/mês a R$ 40.000/mês') &
    (dados['SALARIO'] > limite_superior),
    'SALARIO'
] = media_30_40
# Aqui atribuímos a ele a media do 30 40 substituindo os valores
```

🔍 1. Filtragem das linhas: python Copiar Editar (dados['FAIXA SALARIAL'] == 'de R 30.001/mêsaR  40.000/mês') & (dados['SALARIO'] > limite_superior) 👉 Aqui, o código está buscando as pessoas que:

Declararam que ganham entre R 30.001eR  40.000 por mês;

Mas que na coluna de salário real ('SALARIO') têm um valor acima do limite aceitável (um outlier).

🎯 2. Seleciona a coluna SALARIO dessas linhas: python Copiar Editar dados.loc[ ..., 'SALARIO'] 👉 O .loc localiza somente a coluna SALARIO das linhas que foram filtradas acima.

🛠️ 3. Substituição pelo valor médio: python Copiar Editar = media_30_40 👉 Aqui, está dizendo:

"Substitua o valor de salário dessas pessoas pela média dos salários da faixa 'de R 30.001aR  40.000'."

Ou seja, está tratando os valores fora do esperado e deixando mais coerente com o restante dos dados.

✅ Resumo: Você está limpando dados extremos (outliers) em uma faixa salarial específica, substituindo valores fora do padrão por uma média realista, para garantir uma análise mais confiável.

### Featuring engineering

🔹 Selecionar,

🔹 Criar,

🔹 Transformar ou combinar colunas (ou variáveis) do seu conjunto de dados, para melhorar o desempenho de modelos de Machine Learning.

💡 Exemplo simples: Imagina que você tem uma coluna com a data de nascimento. Durante o feature engineering, você pode:

Criar uma nova coluna chamada "idade"

Categorizar a idade como: jovem, adulto, idoso

Criar variáveis binárias (0 e 1) para representar categorias

Tudo isso é feito para alimentar melhor um modelo de IA ou estatístico.

```python
# Fazer uma função que vai receber os dados da coluna gestor e da coluna nível, que retorna se apessoa é gestora , ou se tem um valor de nível já definido
def preencher_nivel(gestor, nivel):
    if gestor == 1:
        return "Pessoa Gestora"
    else:
        return nivel
```

```python
#Grafico para ver se o salário varia de acordo com idade,  com matiplotlib
salario_por_idade = dados.groupby('IDADE')['SALARIO'].mean()
```

```python
salario_por_idade
```

```python
plt.figure()
plt.plot(salario_por_idade.index, salario_por_idade.values, marker = "o", linestyle ="--")
plt.xlabel('Idade')
plt.ylabel('Média salário Médio')
plt.title('Média de salário por Idade')
plt.grid(True)
plt.show
```
